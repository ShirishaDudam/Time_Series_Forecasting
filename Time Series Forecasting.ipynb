```python
from statsmodels.tsa.stattools import adfuller
def test_stationarity(data):
    results= adfuller(data)
    print("Null Hypotheis: There is no Stationarity")
    print("Alternate Hypotheis: There is Stationarity")
    print('ADF statistics:', results[0])
    print('p_value:', results[1])
    print('Critical values')
    for key, val in results[4].items():
        print(key, val)
    if results[1]<0.05:
        print("Reject the Null Hypothesis: Data is Stationary")
    else:
        print("Accept the Null Hypothesis: Data is Non-Stationary")

    rolling_mean= data.rolling(window=6).mean()
    rolling_std= data.rolling(window= 6).std()
    plt.subplot(3,1,2)
    plt.plot(data, color='blue', label='original')
    plt.plot(rolling_mean, color='red', label='Rolling Mean')
    plt.legend()

    plt.subplot(3,1,3)
    plt.plot(rolling_std, color='yellow', label='Rolling std')
    plt.tight_layout()
    plt.show()
    
test_stationarity(train)   
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
train =train.asfreq("D")
train=train.fillna(method='ffill')
log_data=np.log(train)
ep=1e-10
decomposition=seasonal_decompose(log_data, model='additive')
```

```python
trend=decomposition.trend
seasonal=decomposition.seasonal
residual=decomposition.resid
```

```python
plt.subplot(4,1,1)
plt.plot(train, color='orange', label='Original')
plt.legend(loc='best')

plt.subplot(412)
plt.plot(trend, color='blue', label='Trend')
plt.legend(loc='best')

plt.subplot(413)
plt.plot(seasonal, color='green', label='seasonal')
plt.legend(loc= 'best')


plt.subplot(414)
plt.plot(residual, color='pink', label='residual')
plt.legend(loc= 'best')

plt.show()
```

```python
log_diff= log_data- log_data.shift()
roll_mean= log_diff.rolling(12).mean()
roll_std=log_diff.rolling(12).std()

plt.figure(figsize=(10,4))
plt.plot(log_diff, color='blue', label= 'log diff')
plt.plot(roll_mean, color='red', label='roll mean')
plt.plot(roll_std, color='yellow', label='roll std')
plt.legend()
plt.show()
```

```python
fig= plt.figure(figsize=(10,8))
ax1= fig.add_subplot(211)
fig= sm.graphics.tsa.plot_acf(log_diff.dropna(), lags= 40, ax=ax1)

ax2=fig.add_subplot(212)
fig= sm.graphics.tsa.plot_pacf(log_diff.dropna(), lags=40, ax= ax2)

plt.show()
```

```python
from statsmodels.tsa.arima.model import ARIMA
plt.figure(figsize=(10,8))

model= ARIMA(log_diff, order=(2,2,2)) #p d q
result_arima= model.fit()
plt.plot(log_diff)
plt.plot(result_arima.fittedvalues, color='red')
plt.show()
```

```python
arima_diff_pred=pd.Series(result_arima.fittedvalues, copy=True)
arima_culumative= arima_diff_pred.cumsum()
arima_log= pd.Series(log_data.iloc[0], index= log_data.index)
arima_log= arima_log.add(arima_culumative, fill_value=0)
pred_arima=np.exp(arima_log)

plt.figure(figsize=(10,7))
plt.plot(train, color='blue', label= 'original')
plt.plot(pred_arima, color='orange', label='predicted')
plt.legend()
plt.title('RMSE:' +str(np.sqrt((sum(pred_arima-train)**2)/len(train))) )
plt.show()
```
